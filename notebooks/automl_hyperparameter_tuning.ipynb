```python
# load the ready_sic_sst_data.zip file and extract the data

import zipfile
import os

zip_file_path = '/Users/skylargale/Desktop/vscode/MLGEO2024_SeaIcePrediction/data/ready_sic_sst_data.zip'

if os.path.exists(zip_file_path):
    # unzip the file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall('.')

    print('Data extracted successfully!')
else:
    print(f'File not found: {zip_file_path}')
```

```python
# load the .nc file
import xarray as xr

data_path = '/Users/skylargale/Desktop/vscode/MLGEO2024_SeaIcePrediction/data/ready_sic_sst_data.nc'
ds = xr.open_dataset(data_path)

ds = ds.isel(latitude=slice(0, 81))
ds
```

```python
# use pycaret to train and select models
from pycaret.regression import *

# remove rows with missing values in the target column 'sic'
df = ds.to_dataframe().reset_index()
df = df.dropna(subset=['sst'])

# setup the data
s = setup(data=df, target='sst')

# compare models
best = compare_models()

# tune the best model
tuned_best = tune_model(best)

# plot the model
plot_model(tuned_best, plot='error')

# save the model
save_model(tuned_best, 'sst_model')

# load the model
loaded_model = load_model('sst_model')

# make predictions
predictions = predict_model(loaded_model, df)

print(predictions)
```
