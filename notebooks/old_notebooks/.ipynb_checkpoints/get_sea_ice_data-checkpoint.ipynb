```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import os

# Home path - edit this to the directory you cloned the repo to
home = '/glade/u/home/skygale/'

# Step 1: Create a directory for storing downloaded .nc files
raw_data_dir = home + 'MLGEO2024_SeaIcePrediction/data/raw/seaice_conc'
# Only make the directory if it does not already exist
os.makedirs(raw_data_dir, exist_ok=True)

# Step 2: Scrape the website for .nc file links using BeautifulSoup
url = "https://noaadata.apps.nsidc.org/NOAA/G02202_V4/north/monthly/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Step 3: Find all links ending with .nc
nc_links = []
for link in soup.find_all('a'):
    href = link.get('href')
    if href and href.endswith('.nc'):
        nc_links.append(url + href)  # Construct the full URL

# Step 4: Download the .nc files into the data directory
for nc_url in nc_links:
    filename = os.path.join(raw_data_dir, nc_url.split('/')[-1])
    response = requests.get(nc_url)
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f'Downloaded {filename}')
```
